```python
import matplotlib.pyplot as plt
import glob
```

```python
image_names = ['bg.jpg', 'fg.jpg', 'fg_mask.jpg', 'fg_bg.jpg', 'fg_bg_mask.jpg', 'fg_bg_depth.jpg']
```

```python
bg_paths = sorted(glob.glob('./bg/*'))
fg_paths = sorted(glob.glob('./fg/*'))
fg_mask_paths = sorted(glob.glob('./fg_mask/*'))
fg_bg_paths = sorted(glob.glob('./fg_bg/*'))
fg_bg_mask_paths = sorted(glob.glob('./fg_bg_mask/*'))
fg_bg_depth_paths = sorted(glob.glob('./fg_bg_depth/*'))

all_paths = [bg_paths, fg_paths, fg_mask_paths, fg_bg_paths, fg_bg_mask_paths, fg_bg_depth_paths]
```

```python
for index, each_path in enumerate(all_paths):
    img_name = image_names[index]
    plt.figure(figsize=(20, 2))
    for num, path in enumerate(each_path, start=1):
        img = plt.imread(path)
        plt.subplot(1, 10, num)
        plt.imshow(img, cmap='gray')
        plt.axis('off')
    plt.subplots_adjust(left=0, bottom=0, wspace=0, hspace=0)
    plt.savefig(img_name, bbox_inches='tight')
```
